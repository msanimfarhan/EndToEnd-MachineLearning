```python
#decision boundary 
#sigmoid function 1 / 1+ e ^(-z) 

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn  as sns  
```

```python
def segmoid(x):
    return 1 / (1 + np.exp(-x)) 
```

```python
z = np.arange(-10,11,0.1) 
values = segmoid(z) 

plt.plot(z,values)
```
